```python
import pandas as pd
import random

# CSV 파일 경로
csv_file_path = '/content/drive/MyDrive/subject CSV/ADNI3_M_MCI.csv'

# CSV 파일 읽기
df = pd.read_csv(csv_file_path)

flair_rows = df[df['Description'] == 'Sagittal 3D FLAIR']

# 무작위로 50개의 행 선택
random_sample = flair_rows.sample(n=50, random_state=42)  # random_state는 재현성을 위한 시드값입니다.

# 선택된 행을 새로운 CSV 파일로 저장
output_csv_file_path = '/content/drive/MyDrive/subject CSV/ADNI3_M_MCI_50.csv'
random_sample.to_csv(output_csv_file_path, index=False)
```
