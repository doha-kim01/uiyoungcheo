```python
import numpy as np
import nibabel as nib  # You'll need this library to work with NIfTI MRI images

# Load the MRI image using nibabel (replace 'your_image.nii' with your image file)
img = nib.load('your_image.nii')

# Extract the image data as a NumPy array
image_data = img.get_fdata()

# Find the maximum intensity value in the entire image
max_intensity = np.max(image_data)

# Normalize the image data by dividing each voxel value by the max intensity
normalized_image = image_data / max_intensity

# Save the normalized image to a new NIfTI file (optional)
normalized_img = nib.Nifti1Image(normalized_image, img.affine)
nib.save(normalized_img, 'normalized_image.nii')

# Now, 'normalized_image' contains the voxel-wise intensity values between 0 and 1
```

```python
import os
import numpy as np
import nibabel as nib

# Specify the folder containing the NIfTI files
folder_path = '/content/drive/MyDrive/2023 의학영상처리/MRI/MCI/M'

# List all files in the folder
file_list = os.listdir(folder_path)

# Loop through each file in the folder
for file_name in file_list:
    if file_name.endswith('.nii') or file_name.endswith('.nii.gz'):  # Check if the file is a NIfTI file
        # Construct the full path to the NIfTI file
        file_path = os.path.join(folder_path, file_name)

        # Load the MRI image using nibabel
        img = nib.load(file_path)

        # Extract the image data as a NumPy array
        image_data = img.get_fdata()

        # Find the maximum intensity value in the entire image
        max_intensity = np.max(image_data)

        # Normalize the image data by dividing each voxel value by the max intensity
        normalized_image = image_data / max_intensity

        # Save the normalized image to a new NIfTI file (optional)
        normalized_img = nib.Nifti1Image(normalized_image, img.affine)
        normalized_file_path = os.path.join(folder_path, 'normalized_' + file_name)
        nib.save(normalized_img, normalized_file_path)

        print(f"Normalization completed for {file_name}. Normalized file saved as {normalized_file_path}")
```
